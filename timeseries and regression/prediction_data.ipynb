```python
#create PredictionData object
path = "E:\Courses\Python\project 2\California house prices.csv"
target = "House_price"
pred = PredictionData(path, target)
```

Scatter plot of all the features

```python
pred.scatter_plot()
```

Median Income seems to have the best correlation with te house prices while for other variables a correlation is not apparent from the scatter plot.

Standardize the data so that it has a mean of zero and a standard deviation of one

```python
data_norm = pred.normalize()
```

### Linear regression models to establish the correlation between all the features and the target

```python
feature, R, beta = pred.linear_reg(data_norm)
```

### Best Feature

```python
for index, f in enumerate(feature):
    if f != target:
        print("Feature: ", f ,", Coefficients[c, m]: ", beta[index], ", Residual: ", R[index])    
```


##### Best model we obtain based on the residuals is: MedInc. This makes sense intuitively as we can expect the median income of the block to be correlated to the house prices in the block.
